```python
import pandas as pd
data=pd.read_csv(r'Employee data.csv')
```

Q1. Generate a histogram using any library to visualize the distribution of salaries among employees in the dataset.

```python
#Using seaborn Library

import seaborn as sns
import matplotlib.pyplot as plt

salary_data = data['salary']

sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.histplot(salary_data, bins=10, kde=False, color='skyblue')
plt.title('Distribution of Salaries')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.show()
```

```python
import sweetviz as sv
employee_report = sv.analyze(data)
employee_report.show_html('employee.html')
```

Q2. Generate a bar plot to compare the average salary of male and female employees using any library.

```python
#Using seaborn
sns.set(style="whitegrid")
plt.figure(figsize=(8, 6))
sns.barplot(x='gender', y='salary', data=data, ci=None, palette='pastel')
plt.title('Average Salary by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Salary')
plt.show()
```

```python
import dtale
d=dtale.show(data)
d.open_browser()
```

Q3. Create a scatter plot using any library to illustrate the relationship between previous work experience (prevexp) and the current salary of employees.


```python
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.scatterplot(x='prevexp', y='salary', data=data, color='skyblue')
plt.title('Relationship between Previous Work Experience and Salary')
plt.xlabel('Previous Work Experience')
plt.ylabel('Salary')
plt.show()
```

```python
import dtale
d=dtale.show(data)
d.open_browser()
```

Q4. Generate a pie chart to visualize the distribution of educational backgrounds among employees.

```python
education_counts = data['educ'].value_counts()

plt.figure(figsize=(8, 8))
sns.set_palette('Pastel1')
plt.pie(education_counts, labels=education_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Educational Backgrounds')
plt.show()
```
